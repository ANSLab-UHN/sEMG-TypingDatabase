```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import os
import datetime
import csv
from tabulate import tabulate
```

```python
plt.rcParams.update({
    'axes.titlesize': 18,  # Title font size
    'axes.labelsize': 18,  # Label font size
    'xtick.labelsize': 18, # X-axis tick font size
    'ytick.labelsize': 18  # Y-axis tick font size
})
```

```python
file_name = './../ClassificationResults/ResultsSummary.xlsx'
FeatureResults = pd.read_excel(io=file_name, sheet_name='Features')
CrossTestResults = pd.read_excel(io=file_name, sheet_name='CrossTest')
CrossParticipantResults = pd.read_excel(io=file_name, sheet_name='CrossParticipant')
WindowResults = pd.read_excel(io=file_name, sheet_name='Window')
print(FeatureResults.head(5))  # print first 5 rows of the dataframe
```

```python
sns.barplot(x=FeatureResults['Features'], y=FeatureResults['CA'],yerr=FeatureResults['SD'])
plt.xticks(rotation=55,ha='right')
plt.ylim(0.75, 1)
# Add labels and title
plt.xlabel('Feature Sets')
plt.ylabel('Classification Accuracy')
plt.title('Classification Accuracy for Best Feature Sets')
#plt.gcf().set_size_inches(10, 6)  # Adjust figure size
plt.subplots_adjust(left=-0.5) 
plt.savefig('FeatureBarplot.pdf', format='pdf', bbox_inches='tight')
# Show plot
plt.show()
```

```python
sns.barplot(x=FeatureResults['Name'], y=FeatureResults['CA'],yerr=FeatureResults['SD'],label=(FeatureResults['Name']+": "+FeatureResults['Features']))
plt.xticks(rotation=55,ha='right')
plt.ylim(0.75, 1)
# Add labels and title
plt.xlabel('Feature Sets')
plt.ylabel('Classification Accuracy')
plt.title('Classification Accuracy for Best Feature Sets')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left',fontsize=16)
# bbox = [1.05,0,1,1]  # [left, bottom, width, height]
# table = plt.table(cellText=FeatureNames.values, colLabels=FeatureNames.columns,bbox=bbox,loc='right',cellLoc='left')
# table.auto_set_column_width([0, 1])
# table.auto_set_font_size(False)
# table.set_fontsize(12)
# table.scale(1.2, 1.2)

plt.gcf().set_size_inches(10, 8)  # Adjust figure size
plt.subplots_adjust(left=-0.3) 

plt.savefig('FeatureBarplot_Legend.pdf', format='pdf', bbox_inches='tight')
# Show plot
plt.show()
```

```python
# Create a figure and axis
fig, ax = plt.subplots()
# Hide axes
ax.axis('off')
# Create table
FeatureNames = FeatureResults.iloc[:,0:2]
table = ax.table(cellText=FeatureNames.values, colLabels=FeatureNames.columns, loc='center',cellLoc='left')
# Modify cell height and width
table.auto_set_column_width([0, 1])
table.auto_set_font_size(False)
table.set_fontsize(12)
table.scale(1.2, 1.2)
plt.savefig('FeatureTable.pdf', format='pdf', bbox_inches='tight')
# Show plot
plt.show()
```

```python
sns.scatterplot(data=WindowResults, x='Window length', y='CA',color='red',marker='x',linewidth=2)
for i in range(len(WindowResults)):
    plt.errorbar(WindowResults['Window length'][i], WindowResults['CA'][i], yerr=WindowResults['SD'][i], fmt='none', color='black', capsize=5)
sns.lineplot(data=WindowResults, x='Window length', y='CA',color='blue',linewidth=1.5)
plt.xticks(rotation=0,ha='center')
plt.ylim(0.75, 1)
# Add labels and title
plt.xlabel('Window Length [s]')
plt.ylabel('Classification Accuracy')
plt.title('Classification Across Tested Window Lengths')
#plt.gcf().set_size_inches(10, 6)  # Adjust figure size
plt.subplots_adjust(left=-0.5) 
# Show plot
plt.savefig('WindowCA.pdf', format='pdf', bbox_inches='tight')
plt.show()
```

```python
sns.set_palette('deep')
sns.barplot(x=CrossTestResults['Participant'], y=CrossTestResults['Average'],label=CrossTestResults['Features'],color='blue')
plt.xticks(rotation=0,ha='center')
plt.ylim(0, 0.5)
# Add labels and title
plt.xlabel('Participant')
plt.ylabel('Classification Accuracy')
plt.title('Classification Accuracy for Across Tests')
#plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left',fontsize=16)
# bbox = [1.05,0,1,1]  # [left, bottom, width, height]
# table = plt.table(cellText=FeatureNames.values, colLabels=FeatureNames.columns,bbox=bbox,loc='right',cellLoc='left')
# table.auto_set_column_width([0, 1])
# table.auto_set_font_size(False)
# table.set_fontsize(12)
# table.scale(1.2, 1.2)

#plt.gcf().set_size_inches(10, 6)  # Adjust figure size
plt.subplots_adjust(left=-0.3) 

plt.savefig('CrossTestBarplot.pdf', format='pdf', bbox_inches='tight')
# Show plot
plt.show()
```

```python
sns.barplot(x=CrossParticipantResults['Participant'], y=CrossParticipantResults['CA'])
plt.xticks(rotation=0,ha='center')
plt.ylim(0, 0.5)
# Add labels and title
plt.xlabel('Participant')
plt.ylabel('Classification Accuracy')
plt.title('Classification Accuracy for Across Participants')
#plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left',fontsize=16)
# bbox = [1.05,0,1,1]  # [left, bottom, width, height]
# table = plt.table(cellText=FeatureNames.values, colLabels=FeatureNames.columns,bbox=bbox,loc='right',cellLoc='left')
# table.auto_set_column_width([0, 1])
# table.auto_set_font_size(False)
# table.set_fontsize(12)
# table.scale(1.2, 1.2)

#plt.gcf().set_size_inches(10, 6)  # Adjust figure size
plt.subplots_adjust(left=-0.3) 

plt.savefig('CrossParticipant.pdf', format='pdf', bbox_inches='tight')
# Show plot
plt.show()
```
